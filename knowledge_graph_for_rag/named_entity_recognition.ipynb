```python
import spacy
from spacy.lang.en.examples import sentences 
nlp = spacy.load("en_core_web_sm")
```

```python
from llama_index.core import download_loader

from llama_index.readers.wikipedia import WikipediaReader

loader = WikipediaReader()

documents = loader.load_data(
    pages=["Taylor Swift"], auto_suggest=False
)
```

```python
doc = nlp(documents[0].text)
```

```python
print({(X.text, X.label_) for X in doc.ents if X.label_=='PERSON'})
```

```python
# For testing, we manually copy over all the entities person with the word Taylor in it
prompt = """
Given the following list of named entities and original text, return a deduplicated list of named entities. 

The named entities should be returned delimited by ,

Named entities:
('Taylor', 'PERSON')
('James Taylor', 'PERSON')
('Taylor Swift', 'PERSON')
('Taylor Alison Swift', 'PERSON')
('Taylor Lautner', 'PERSON')

Original text:
""" + documents[0].text
```
